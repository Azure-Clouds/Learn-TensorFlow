### TensorFlow and House Price Prediction

The task here is to predict the house price following a simple formula:<br>
House Cost = 50 K + 50 K * Number of Bedroom. <br>

Train a Neural Net and then We check its performance for arbitrary input. 

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

#### Input Data

First we create two input arrays (x, y) where y follows the formula given above for corresponding x. As you can understand x will be number of bedrooms. 

```python
xs = np.array([1., 2., 3., 4., 6., 8.], dtype='float32') # number of rooms
ys = np.array([2., 3., 4., 5., 7., 9.], dtype='float32')*50e3 # price in dollar
```

Create a function that trains a model for a given number of epochs

```python
def new_price(no_of_rooms): #should be an one dimensional array, following the definition of ys
    input_room = xs
    label_price = ys
    model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    model.compile(optimizer='sgd', loss='mean_squared_error')
    model.fit(xs, ys, epochs=1000)
    return model.predict(no_of_rooms)
```

Let's See ow the network performs, For input we give 5, i.e. no of rooms are 5 

```python
u = new_price([5])
```

```python
print ("price of house with 5 bedrooms:   %d  dollars" % (u[0][0]))
```

As we see the price of the house is very close to our prediction, 300000 dollars
